```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyeo import classification as cls

from tpot import TPOTClassifier

from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
```

```python

band_labels = ["band_{}".format(number) for number in range(1,13)]

training_data = pd.read_csv("data/training_for_SVM_just_data.csv", names=
                           ["class"]+band_labels + ["segment_id"])
```

```python
training_data
```

```python
%matplotlib inline
training_data.groupby(by="class").mean().drop(['segment_id'], axis=1).transpose().plot()
```

```python
def normalise(data):
    return ((data-data.min())/(data.max()-data.min())*np.iinfo(np.uint32).max).fillna(0).astype('uint32')

normalised_bands = training_data.drop(['class', 'segment_id'], axis=1).apply(normalise, axis=0)
normalised_bands.describe()
normalised_bands.dtypes
```

```python
preproc_training = pd.concat([training_data[["class", "segment_id"]], normalised_bands], axis = 1)
label_order = ["class"] + band_labels + ["segment_id"]
preproc_training = preproc_training[label_order]
preproc_training
```

```python
for i in range(1,13):
    preproc_training.plot.scatter(x='class', y='band_{}'.format(i))
```

```python
preproc_training.groupby(by="class").mean().drop(['segment_id'], axis=1).transpose().plot.scatter(x='class', y='band_5')
```

```python
labels = preproc_training['class']
features = preproc_training.loc[:,'band_1':].drop(["segment_id"], axis = 1)
labels
```

```python
f_train, f_test, l_train, l_test = train_test_split(features, labels)
model = SVC(kernel='rbf')
#model.fit(f_train, l_train)
#print(model.score(f_test, l_test))
```
